# List in Python

## Basic Operations on list

List concatenations: You can use the add (+)  operator to concatenate two list  together. It basically appends the second list to the list one.  
S.extend(t): Extend the list S with content of list T.  
List replication: You can use the multiply (*) operator to replicate the items in the list. You have to multiply a list with an integer for the replication.  
List * n : if the value of n is less than 0, it is treated as zero, which yields an empty list.  
The items in the list are not copied, they are referenced multiple times. ( so any change made to one is reflected in the final list).  
If its a list of list, modifying any element of the list, modifies the result list.   
Comparison Operator: you can pass a comparison operator between two list.  
The operator do an element by element comparison between the lists.  
The first items of both the list are compared, if they differ, this determine the outcome of the comparison.  
If all the items of the list (or sequence in general) compares equal, the sequence are considered equal.  
If one sequence is an initial sub-sequence of the other, the shorter sequence is the smaller (lesser) one. 
The exceptions are not suppressed, if any comparison fails, the entire operation will fail. 
Presence of an Element: to check whether or not an element is in list we use the IN/ Not IN operator.  
It returns True or False depending on the presence.  
In/Not In have the same priority as the comparison operator.  
 


## Built-in functions

List. Index( X , <start index>, <stop index>):  
Returns the index of the position of the first occurrence of X    
If there are duplicates, it reports the first occurrence only.  
Raises a Value Error if X is not present.  
Append():  
Add the element at the end of the list 
The list is modified in place.  
Insert (Index, Value):  
Add the element at the specified location 
Remove (Value): 
Removes the value from the list - irrespective of its location.  
If there are multiple values, the first instance get removed 
Pop (I): 
Removes the element at index i 
If I is not defined, it removes the last element from the list. 
It is also possible to use list as a queue. 
The first in and fist out using the popleft() function - which removes the left most element from the list.  
To use this feature, use collection.deque  which allows popping from both the ends of the list.  
Del list [<index>]: 
This function is used to delete the element from the list at the defined index.  
Deletes the memory associated with the list.  
Clear(): 
Clear all the content of the list.  
Copy(): 
Creates the copy of the list.  
The copy remain, also if the original list has been deleted. This is called as Deep Copy 
You can create the shallow copy by using the assignment operator. They refer of the same list in the memory (passes the reference of the object). The changes made in the one will be reflected in the other as well. This holds true even in the local scope.  
However if you use the list() function to assign a list, it creates a new list in the memory.  
Sort( reverse = T/F, key = str.lower):
